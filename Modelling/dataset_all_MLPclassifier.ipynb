```python
from sklearn import svm, metrics
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
import numpy as np
```

```python
df = pd.read_csv('./data_all/dataset_completed')
```

```python
del df['Unnamed: 0']
```

```python
df
```

```python
X = df[ df.columns[1:15] ]
X.head()
```

```python
def func(x) :
    if 3000000 > int(x) :
        return 1
   
    if 3000000 <= int(x) < 5000000 :
        return 2
     
    if 5000000 <= int(x):
        return 3
```

```python
Y = df['audience_cnt']
```

```python
Y = Y.apply(func)
```

```python
X_train, X_test, Y_train, Y_test= train_test_split( X, Y, test_size=.2 )
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
mlp_multilabel = MLPClassifier(hidden_layer_sizes=( 300,100), max_iter=10000, 
                               random_state=42).fit(X_train, Y_train)
```

```python
mlp_multilabel.score(X_test, Y_test)
```

```python
col = X.columns
```

```python
Y_pred = mlp_multilabel.predict(X_test)
Y_pred
```

```python
np.sum(Y_test.astype(int) == Y_pred)/Y_test.shape[0]
```

```python
pre = mlp_multilabel.predict(X_test)
```

```python
acSore = metrics.accuracy_score(Y_test,pre)
acSore
```

```python
print(metrics.classification_report(Y_test, pre))
```
