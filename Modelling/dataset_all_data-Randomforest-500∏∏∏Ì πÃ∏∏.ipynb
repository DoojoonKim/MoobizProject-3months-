```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df = pd.read_csv('./data_all/dataset_completed')
df
```

```python
del df['Unnamed: 0']
```

```python
df.set_index('movies',inplace=True)
```

```python
df
```

```python
X = df[ df.columns[0:14] ]
X.head()
```

```python
y = df['audience_cnt']
y.head()
```

```python
def func(x) :
    if 5000000 > int(x) :
        return 0
    else :
        return 1
```

```python
y = y.apply(func)
```

```python
X_train, X_test, y_train, y_test= train_test_split( X, y, test_size=.2 )
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = [
    {'n_estimators':[3,10,30,50,60,70,100], 'max_features':[2,4,6]},
    {'bootstrap':[False], 'n_estimators':[3,10,30,50,60,70,100], 'max_features':[2,3,4]},
]
forest_reg = RandomForestClassifier()
grid_search = GridSearchCV(forest_reg, param_grid,cv = 5,
                         scoring='neg_mean_squared_error',
                         return_train_score=True)
grid_search.fit(X,y)
grid_search.best_params_
```

```python
clf1 = RandomForestClassifier(max_features=2,n_estimators=70)
clf1.fit(X_train, y_train)
```

```python
col = X.columns
for i in range(len(X.columns)):
    print(col[i],clf1.feature_importances_[i])
```

```python
pre = clf1.predict(X_test)
acSore = metrics.accuracy_score(y_test,pre)
acSore
```

```python
print(metrics.classification_report(y_test, pre))
```
