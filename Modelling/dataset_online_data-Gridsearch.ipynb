```python
import pandas as pd
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df1 = pd.read_csv('./data_all/online_data_all')
df2 = pd.read_csv('./data_2012/online_data_2012')
df3 = pd.read_csv('./data_2011/online_data_2011')
df4 = pd.read_csv('./data_2010/online_data_2010')
df5 = pd.read_csv('./data_2009/online_data_2009')
df6 = pd.read_csv('./data_2008/online_data_2008')
```

```python
df = pd.concat([df1,df2,df3,df4,df5,df6])
```

```python
df.drop_duplicates('movies',keep='first')
```

```python
df.set_index('movies',inplace=True)
```

```python
df
```

```python
X = df[ df.columns[1:7] ]
X.head()
```

```python
y = df['audience_cnt']
y.head()
```

```python
def func(x) :
    if 3000000 > int(x) :
        return 1
   
    if 3000000 <= int(x) < 5000000 :
        return 2
     
    if 5000000 <= int(x):
        return 3
```

```python
y = y.apply(func)
```

```python
X_train, X_test, y_train, y_test= train_test_split( X, y, test_size=.2 )
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
pipe = Pipeline( [ ('preprocessing',StandardScaler() ),
                   ('classifier',SVC() ) ] )
```

```python
param = [
    {
        'classifier':[SVC()],
        'classifier__C':[0.001, 0.01, 1, 10, 100],
        'classifier__gamma':[0.001, 0.01, 1, 10, 100],
        'preprocessing':[StandardScaler()]
    },
    {
        'classifier':[RandomForestClassifier(n_estimators=100)],
        'classifier__max_features':[1,2,3],
        # 전처리기 사용 않함
        'preprocessing':[None]
    }
]
```

```python
grid = GridSearchCV( pipe, param, cv=5)
grid.fit( X_train, y_train )
print( grid.best_params_ )
print( grid.best_score_ )
print( grid.score(X_test, y_test) )
```
