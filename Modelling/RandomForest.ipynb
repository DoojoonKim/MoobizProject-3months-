```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df = pd.read_csv('./data_all/online_data_all')
```

```python
X = df[ df.columns[1:8] ]
X
```

```python
Y = df['audience_cnt']
```

```python
def func(x) :
    if 3000000 > int(x) :
        return 1
   
    if 3000000 <= int(x) < 5000000 :
        return 2
     
    if 5000000 <= int(x):
        return 3
```

```python
Y = Y.apply(func)
Y
```

```python
X_train, X_test, Y_train, Y_test= train_test_split( X, Y, test_size=.8 )
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
clf = RandomForestClassifier(n_estimators=10)
clf.fit(X_train, Y_train)
```

```python
col = X.columns
```

```python
for i in range(len(X.columns)):
    print(col[i],clf.feature_importances_[i])
```

```python
pre = clf.predict(X_test)
```

```python
acSore = metrics.accuracy_score(Y_test,pre)
acSore
```

```python
print(metrics.classification_report(Y_test, pre))
```

```python
np.sum(Y_test.astype(int) == pre)/Y_test.shape[0]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
pipe = Pipeline( [ ('preprocessing',StandardScaler() ),
                   ('classifier',SVC() ) ] )
```

```python
param = [
    {
        'classifier':[SVC()],
        'classifier__C':[0.001, 0.01, 1, 10, 100],
        'classifier__gamma':[0.001, 0.01, 1, 10, 100],
        'preprocessing':[StandardScaler()]
    },
    {
        'classifier':[RandomForestClassifier(n_estimators=100)],
        'classifier__max_features':[1,2,3],
        # 전처리기 사용 않함
        'preprocessing':[None]
    }
]
```

```python
grid = GridSearchCV( pipe, param, cv=5)
grid.fit( X_train, Y_train )
print( grid.best_params_ )
print( grid.best_score_ )
print( grid.score(X_test, Y_test) )
```
