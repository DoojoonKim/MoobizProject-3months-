```python
import pandas as pd
from sklearn import tree
```

```python
df = pd.read_csv('./data_all/dataset_completed')
df
```

```python
del df['Unnamed: 0']
```

```python
X = df[ df.columns[1:15] ]
X
```

```python
y = df['audience_cnt']
```

```python
def func(x) :
    if 3000000 > int(x) :
        return 1
   
    if 3000000 <= int(x) < 5000000 :
        return 2
     
    if 5000000 <= int(x):
        return 3
```

```python
y = y.apply(func)
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
X_train, X_test, y_train, y_test= train_test_split( X, y, test_size=.2 )
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
sc = StandardScaler()
sc.fit(X_train)
```

```python
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
movie_tree = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
movie_tree.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = movie_tree.predict(X_test)
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred_tr))
```

```python
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image
```

```python
dot_data = export_graphviz(movie_tree, out_file=None, feature_names=['rank','audiCnt','showCnt','NationAtt','RatingAtt','openDt_season','openDt_peak','grade','p_breviews','n_breviews','breviews','p_reviews','n_reviews','reviews'],
                          class_names=['1','2','3'], filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
```
