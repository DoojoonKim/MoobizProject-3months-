```python
import pandas as pd
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df = pd.read_csv('./data_all/online_data_all')
```

```python
df.drop_duplicates('movies',keep='first',inplace=True)
```

```python
df.set_index('movies',inplace=True)
```

```python
df
```

```python
X = df[ df.columns[0] ]
X.head()
```

```python
y = df['audience_cnt']
y.head()
```

```python
def func(x) :
    if 3000000 > int(x) :
        return 1
   
    if 3000000 <= int(x) < 5000000 :
        return 2
     
    if 5000000 <= int(x):
        return 3
```

```python
y = y.apply(func)
```

```python
X_train, X_test, y_train, y_test= train_test_split( X, y, test_size=.2 )
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)
```

```python
col = X.columns
for i in range(len(X.columns)):
    print(col[i],clf.feature_importances_[i])
```

```python
pre = clf.predict(X_test)
acSore = metrics.accuracy_score(y_test,pre)
acSore
```

```python
print(metrics.classification_report(y_test, pre))
```
