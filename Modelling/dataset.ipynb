```python
import pandas as pd
df = pd.read_csv('./data_all/online_data_all')
```

```python
df.head()
```

```python
df['p_percent'] = df['p_reviews']/df['reviews']
```

```python
dfTest = df
```

```python
from sklearn import preprocessing
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
def scaleColumns(df, cols_to_scale):
    for col in cols_to_scale:
        df[col] = pd.DataFrame(min_max_scaler.fit_transform(pd.DataFrame(dfTest[col])),columns=[col])
    return df
```

```python
scaled_df = scaleColumns(dfTest,['grade','p_breviews','n_breviews','breviews','p_reviews','n_reviews','reviews',])
scaled_df
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
X = df[ df.columns[1:8] ]
X
```

```python
Y = df['audience_cnt']
```

```python
def func(x) :
    if 3000000 > int(x) :
        return 1
   
    if 3000000 <= int(x) < 6000000 :
        return 2
     
    if 6000000 <= int(x) :
        return 3
```

```python
Y = Y.apply(func)
Y
```

```python
X_train, X_test, Y_train, Y_test= train_test_split( X, Y, test_size=.8 )
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
clf = RandomForestClassifier(n_estimators=10)
clf.fit(X_train, Y_train)
```

```python
col = X.columns
```

```python
for i in range(len(X.columns)):
    print(col[i],clf.feature_importances_[i])
```

```python
pre = clf.predict(X_test)
```

```python
acSore = metrics.accuracy_score(Y_test,pre)
acSore
```

```python
print(metrics.classification_report(Y_test, pre))
```

```python
x = np.array(pre)
unique, counts = np.unique(x, return_counts=True)

print(np.asarray((unique, counts)).T)
```
