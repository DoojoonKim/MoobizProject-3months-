```python
import pandas as pd
grade_2012 = pd.read_csv('./data_2012/2012_grade')
n_breviews_2012 = pd.read_csv('./data_2012/2012_n_breviews')
p_breviews_2012 = pd.read_csv('./data_2012/2012_p_breviews')
n_reviews_2012 = pd.read_csv('./data_2012/2012_n_reviews')
p_reviews_2012 = pd.read_csv('./data_2012/2012_p_reviews')
audience_cnt_2012 = pd.read_csv('./data_2012/2012_audience_cnt')
b_reviews_2012 = pd.merge(p_breviews_2012,n_breviews_2012)
reviews_2012 = pd.merge(p_reviews_2012,n_reviews_2012)
b_reviews_2012['breviews'] = b_reviews_2012['p_breviews'] + b_reviews_2012['n_breviews']
reviews_2012['reviews'] = reviews_2012['p_reviews'] + reviews_2012['n_reviews']
df1 = pd.merge(b_reviews_2012,reviews_2012)
online_data_2012 = pd.merge(grade_2012,df1)
online_data_2012 = pd.merge(online_data_2012,audience_cnt_2012)
online_data_2012 = online_data_2012.set_index('Unnamed: 0')
online_data_2012.index.name = 'movies'
online_data_2012
```
