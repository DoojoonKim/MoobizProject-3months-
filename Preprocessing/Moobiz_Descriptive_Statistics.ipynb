```python
df_nation = pd.read_excel('./data/국적비교.xlsx')
df_nation
```

```python
df_a = pd.merge(df_nation, df_dataset2, on=('movieNm'))
df_a = df_a.drop_duplicates('movieNm', keep='first')
df_a
```

```python

Nation_var= []
for idx, a in enumerate(df_a['Nation_x']):
    if type(a) is not str:
        continue
    elif a == '한국':
         Nation_var.append(0)
    elif a == '미국':
         Nation_var.append(1)
    else:
         Nation_var.append(2)
df_a['Nation_att'] = pd.Series(Nation_var)
```

```python
df_a['Nation_att'] = pd.Series(Nation_var)
df_a
```

```python
df_a.Rate.describe()
```

```python
Rate_var = []

for idx, a in enumerate(df_a['Rate']):
    if type(a) is not str:
        continue
    elif a == '12세이상관람가':
         Rate_var.append(0)
    elif a == '15세이상관람가':
         Rate_var.append(1)
    elif a == '청소년관람불가':
         Rate_var.append(2)
    else:
         Rate_var.append(3)
df_a['Rate_att'] = pd.Series(Rate_var)
df_a
```
