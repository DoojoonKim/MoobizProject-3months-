```python
import pandas as pd
grade_2010 = pd.read_csv('./data_2010/2010_grade')
n_breviews_2010 = pd.read_csv('./data_2010/2010_n_breviews')
p_breviews_2010 = pd.read_csv('./data_2010/2010_p_breviews')
n_reviews_2010 = pd.read_csv('./data_2010/2010_n_reviews')
p_reviews_2010 = pd.read_csv('./data_2010/2010_p_reviews')
audience_cnt_2010 = pd.read_csv('./data_2010/2010_audience_cnt')
b_reviews_2010 = pd.merge(p_breviews_2010,n_breviews_2010)
reviews_2010 = pd.merge(p_reviews_2010,n_reviews_2010)
b_reviews_2010['breviews'] = b_reviews_2010['p_breviews'] + b_reviews_2010['n_breviews']
reviews_2010['reviews'] = reviews_2010['p_reviews'] + reviews_2010['n_reviews']
df1 = pd.merge(b_reviews_2010,reviews_2010)
online_data_2010 = pd.merge(grade_2010,df1)
online_data_2010 = pd.merge(online_data_2010,audience_cnt_2010)
online_data_2010 = online_data_2010.set_index('Unnamed: 0')
online_data_2010.index.name = 'movies'
online_data_2010
```

```python
online_data_2010.to_csv('./data_2010/online_data_2010')
```
