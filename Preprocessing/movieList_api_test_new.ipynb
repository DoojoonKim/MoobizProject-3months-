```python
def movieCode(movieNm, directorNm):
    req = urllib.request
    url = parse.urlparse('http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=%s&movieNm=%s&directorNm=%s'%(key2,movieNm,directorNm)) 
    query = parse.parse_qs(url.query)
    option = parse.urlencode(query, encoding='UTF-8', doseq=True)
    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?%s'%option
    moviePage = req.urlopen(url)
    movieData = json.loads(moviePage.read())
#     print(movieData)
    movieCd = movieData['movieListResult']['movieList'][0]['movieCd']
    return movieCd
```

```python
def movieInfo(movieCd):
    req = urllib.request
    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key=%s&movieCd=%s'%(key2, movieCd)
    moviePage = req.urlopen(url)
    movieData = json.loads(moviePage.read())
    movie_director = movieData['movieInfoResult']['movieInfo']['directors'][0]['peopleNm']
    movie_actor = []
    movie_act = movieData['movieInfoResult']['movieInfo']['actors']
    for i in range(0, len(movie_act)):
        movie_actor.append( movieData['movieInfoResult']['movieInfo']['actors'][i]['peopleNm'])
    movieCp = (movieData['movieInfoResult']['movieInfo']['companys'])
    for i in range(0, len(movieCp)):
        if movieData['movieInfoResult']['movieInfo']['companys'][i]['companyPartNm']=='배급사':
            film_distributor = movieData['movieInfoResult']['movieInfo']['companys'][i]['companyNm']
    
    return movie_director,movie_actor,film_distributor
```

```python
def actCode(MoInfo):
    req = urllib.request
    actNm = []
    aFilmoNm = []
    for i in range(0, 2):
        url = parse.urlparse('http://www.kobis.or.kr/kobisopenapi/webservice/rest/people/searchPeopleList.json?key=%s&peopleNm=%s'%(key1, MoInfo[i]))
        query = parse.parse_qs(url.query)
        option = parse.urlencode(query, encoding='UTF-8', doseq=True)
        url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/people/searchPeopleList.json?%s'%option
        peoplePage = req.urlopen(url)
        peopleData = json.loads(peoplePage.read())
#         print(peopleData)
        filmoName = peopleData['peopleListResult']['peopleList'][0]['filmoNames']
#         print(filmoName)
        aFilmoNm.append(filmoName.split('|'))
        peopleNm = peopleData['peopleListResult']['peopleList'][0]['peopleNm']
        actNm.append(peopleNm)
    filmoList = dict(zip(actNm, aFilmoNm))
    return filmoList
```

```python
def distributor(MoInfo):
    req = urllib.request
    dFilmo = []
    url = parse.urlparse('http://kobis.or.kr/kobisopenapi/webservice/rest/company/searchCompanyList.json?key=%s&companyNm=%s'%(key1, MoInfo))
    query = parse.parse_qs(url.query)
    option = parse.urlencode(query, encoding='UTF-8', doseq=True)
    url = 'http://kobis.or.kr/kobisopenapi/webservice/rest/company/searchCompanyList.json?%s'%option
    companyPage = req.urlopen(url)
    companyData = json.loads(companyPage.read())
#     print(companyData)
    filmoName = str(companyData['companyListResult']['companyList'][0]['filmoNames'])
    companyName = companyData['companyListResult']['companyList'][0]['companyNm']
    dFilmo.append(filmoName.split(','))
    return companyName, dFilmo
```
