```python
import pandas as pd
grade_2014 = pd.read_csv('./data_2014/2014_grade')
n_breviews_2014 = pd.read_csv('./data_2014/2014_n_breviews')
p_breviews_2014 = pd.read_csv('./data_2014/2014_p_breviews')
n_reviews_2014 = pd.read_csv('./data_2014/2014_n_reviews')
p_reviews_2014 = pd.read_csv('./data_2014/2014_p_reviews')
audience_cnt_2014 = pd.read_csv('./data_2014/2014_audience_cnt')
b_reviews_2014 = pd.merge(p_breviews_2014,n_breviews_2014)
reviews_2014 = pd.merge(p_reviews_2014,n_reviews_2014)
b_reviews_2014['breviews'] = b_reviews_2014['p_breviews'] + b_reviews_2014['n_breviews']
reviews_2014['reviews'] = reviews_2014['p_reviews'] + reviews_2014['n_reviews']
df1 = pd.merge(b_reviews_2014,reviews_2014)
online_data_2014 = pd.merge(grade_2014,df1)
online_data_2014 = pd.merge(online_data_2014,audience_cnt_2014)
online_data_2014 = online_data_2014.set_index('Unnamed: 0')
online_data_2014.index.name = 'movies'
online_data_2014
```

```python
online_data_2014.to_csv('./data_2014/online_data_2014')
```
