```python
import pandas as pd
grade_2016 = pd.read_csv('./data_2016/2016_grade')
n_breviews_2016 = pd.read_csv('./data_2016/2016_n_breviews')
p_breviews_2016 = pd.read_csv('./data_2016/2016_p_breviews')
n_reviews_2016 = pd.read_csv('./data_2016/2016_n_reviews')
p_reviews_2016 = pd.read_csv('./data_2016/2016_p_reviews')
audience_cnt_2016 = pd.read_csv('./data_2016/2016_audience_cnt')
b_reviews_2016 = pd.merge(p_breviews_2016,n_breviews_2016)
reviews_2016 = pd.merge(p_reviews_2016,n_reviews_2016)
b_reviews_2016['breviews'] = b_reviews_2016['p_breviews'] + b_reviews_2016['n_breviews']
reviews_2016['reviews'] = reviews_2016['p_reviews'] + reviews_2016['n_reviews']
df1 = pd.merge(b_reviews_2016,reviews_2016)
online_data_2016 = pd.merge(grade_2016,df1)
online_data_2016 = pd.merge(online_data_2016,audience_cnt_2016)
online_data_2016 = online_data_2016.set_index('Unnamed: 0')
online_data_2016.index.name = 'movies'
online_data_2016
```

```python
online_data_2016.to_csv('./data_2016/online_data_2016')
```
