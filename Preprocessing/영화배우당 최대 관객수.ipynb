```python
import pandas as pd
import os
import sys
```

```python
df_Moobiz_data = pd.read_excel('./data/Moobiz_data.xlsx')
df_Moobiz_data
```

```python
df_actor = pd.read_excel('./data/영화배우당 영화 필모.xlsx')
df_1 = pd.read_excel('./data/배두별 최대 관객수.xlsx')
```

```python
df_actor
```

```python
df_2 = pd.DataFrame({'movieNm':df_Moobiz_data['movieNm'], 'audiCnt': df_Moobiz_data['audiCnt'] })
df_2
```

```python
df_3 = pd.merge(df_2, df_actor, on='movieNm')
df_3
```

```python
df_4=pd.DataFrame({'movieNm':df_3['movieNm'], 'audiCnt':df_3['audiCnt'], 'actor1':df_3['actor1']})
df_4
```

```python
df_4['audiCnt'] = df_4['audiCnt'].sort_values(ascending=False)
# df_4 = df_4.drop_duplicates(subset='movieNm', keep='first')
df_4 = df_4.drop_duplicates(subset='actor1', keep='first')
df_4
```

```python
df_5=pd.DataFrame({'movieNm':df_3['movieNm'], 'audiCnt':df_3['audiCnt'], 'actor2':df_3['actor2']})
df_5
```

```python
df_5['audiCnt'] = df_5['audiCnt'].sort_values(ascending=False)
df_5 = df_5.drop_duplicates(subset='movieNm', keep='first')
df_5 = df_5.drop_duplicates(subset='actor2', keep='first')
df_5
```

```python
df_5.rename(columns={'actor2':'actor1'}, inplace=True)
```

```python
df_5
```

```python
df_6 = pd.concat([df_4,df_5], sort=True)
df_6
```

```python
# df_6.index = pd.RangeIndex(len(df_6))
# df_6['audiCnt'] = df_6['audiCnt'].sort_values(ascending=False)

df_6 = df_6.drop_duplicates(subset='actor1', keep='first')
df_6.rename(columns={'audiCnt':'actorPower'},inplace=True)
df_6
```

```python
df_6.index = pd.RangeIndex(len(df_6))
df_6.to_excel('./data/배우당 최대 관객수.xlsx', encoding='utf-8')
df_6.to_csv('./data/배우당 최대 관객수.csv', encoding='euc-kr')
```
