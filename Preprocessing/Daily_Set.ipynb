```python
import requests
import pandas as pd
import numpy as np
import os
import time

import datetime

import json
from  pandas import DataFrame
```

```python
dir = os.listdir('./data')
dir.sort()
xlsx_list1 = [file for file in dir if file.startswith('searchDay') ]
```

```python
xlsx_list1
```

```python
# df_searchDay_2013 = pd.read_excel('./data/')
for i,data in enumerate(xlsx_list1):
    if i == 0:
        df = pd.read_excel('./data/'+data)

    
    else:
        small_df = pd.read_excel('./data/'+data)
        df = pd.concat([df, small_df]).reset_index(drop = True)
```
