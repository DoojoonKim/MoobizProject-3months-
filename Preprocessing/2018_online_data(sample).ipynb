```python
import pandas as pd
grade_2018 = pd.read_csv('./data_2018/2018_grade')
n_breviews_2018 = pd.read_csv('./data_2018/2018_n_breviews')
p_breviews_2018 = pd.read_csv('./data_2018/2018_p_breviews')
n_reviews_2018 = pd.read_csv('./data_2018/2018_n_reviews')
p_reviews_2018 = pd.read_csv('./data_2018/2018_p_reviews')
audience_cnt_2018 = pd.read_csv('./data_2018/2018_audience_cnt')
b_reviews_2018 = pd.merge(p_breviews_2018,n_breviews_2018)
reviews_2018 = pd.merge(p_reviews_2018,n_reviews_2018)
b_reviews_2018['breviews'] = b_reviews_2018['p_breviews'] + b_reviews_2018['n_breviews']
reviews_2018['reviews'] = reviews_2018['p_reviews'] + reviews_2018['n_reviews']
df1 = pd.merge(b_reviews_2018,reviews_2018)
online_data_2018 = pd.merge(grade_2018,df1)
online_data_2018 = pd.merge(online_data_2018,audience_cnt_2018)
online_data_2018 = online_data_2018.set_index('Unnamed: 0')
online_data_2018.index.name = 'movies'
online_data_2018
```
