```python
df_director_set = pd.DataFrame({'movieNm':df_director['movieNm'], 'director':df_director['director'],
                                'audiCnt': df_director['audiCnt']                   })

director_var = []

for a in df_director['audiCnt']:
#     print(type(a))
#     df_actor_set['actor_var'] = pd.Series(df_actor_set['actor_var'])
#     a = int(a)
    if a < 1000000:
        director_var.append(0)
    elif a >= 1000000 and a < 2000000: 
        director_var.append(1)
    elif a >= 2000000 and a < 3000000: 
        director_var.append(2)
    elif a >= 3000000 and a < 4000000: 
        director_var.append(3)
    elif a >= 4000000 and a < 5000000: 
        director_var.append(4)
    elif a >= 5000000 and a < 6000000: 
        director_var.append(5)
    elif a >= 6000000 and a < 7000000: 
        director_var.append(6)
    elif a >= 7000000 and a < 8000000: 
        director_var.append(7)
    elif a >= 8000000 and a < 9000000: 
        director_var.append(8)
    elif a >= 9000000 and a < 10000000: 
        director_var.append(9)
    elif a >= 10000000: 
        director_var.append(10)
```

```python
df_director_set['director_var'] = pd.Series(director_var)
```

```python
df_director_set.to_csv('./data/감독파워.csv', encoding='euc-kr')
```
