```python
import pandas as pd
grade_2011 = pd.read_csv('./data_2011/2011_grade')
n_breviews_2011 = pd.read_csv('./data_2011/2011_n_breviews')
p_breviews_2011 = pd.read_csv('./data_2011/2011_p_breviews')
n_reviews_2011 = pd.read_csv('./data_2011/2011_n_reviews')
p_reviews_2011 = pd.read_csv('./data_2011/2011_p_reviews')
audience_cnt_2011 = pd.read_csv('./data_2011/2011_audience_cnt')
b_reviews_2011 = pd.merge(p_breviews_2011,n_breviews_2011)
reviews_2011 = pd.merge(p_reviews_2011,n_reviews_2011)
b_reviews_2011['breviews'] = b_reviews_2011['p_breviews'] + b_reviews_2011['n_breviews']
reviews_2011['reviews'] = reviews_2011['p_reviews'] + reviews_2011['n_reviews']
df1 = pd.merge(b_reviews_2011,reviews_2011)
online_data_2011 = pd.merge(grade_2011,df1)
online_data_2011 = pd.merge(online_data_2011,audience_cnt_2011)
online_data_2011 = online_data_2011.set_index('Unnamed: 0')
online_data_2011.index.name = 'movies'
online_data_2011
```

```python
online_data_2011.to_csv('./data_2011/online_data_2011')
```
