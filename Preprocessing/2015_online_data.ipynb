```python
import pandas as pd
grade_2015 = pd.read_csv('./data_2015/2015_grade')
n_breviews_2015 = pd.read_csv('./data_2015/2015_n_breviews')
p_breviews_2015 = pd.read_csv('./data_2015/2015_p_breviews')
n_reviews_2015 = pd.read_csv('./data_2015/2015_n_reviews')
p_reviews_2015 = pd.read_csv('./data_2015/2015_p_reviews')
audience_cnt_2015 = pd.read_csv('./data_2015/2015_audience_cnt')
b_reviews_2015 = pd.merge(p_breviews_2015,n_breviews_2015)
reviews_2015 = pd.merge(p_reviews_2015,n_reviews_2015)
b_reviews_2015['breviews'] = b_reviews_2015['p_breviews'] + b_reviews_2015['n_breviews']
reviews_2015['reviews'] = reviews_2015['p_reviews'] + reviews_2015['n_reviews']
df1 = pd.merge(b_reviews_2015,reviews_2015)
online_data_2015 = pd.merge(grade_2015,df1)
online_data_2015 = pd.merge(online_data_2015,audience_cnt_2015)
online_data_2015 = online_data_2015.set_index('Unnamed: 0')
online_data_2015.index.name = 'movies'
online_data_2015
```

```python
online_data_2015.to_csv('./data_2015/online_data_2015')
```
