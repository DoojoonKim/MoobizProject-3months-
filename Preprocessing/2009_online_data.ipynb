```python
import pandas as pd
grade_2009 = pd.read_csv('./data_2009/2009_grade')
n_breviews_2009 = pd.read_csv('./data_2009/2009_n_breviews')
p_breviews_2009 = pd.read_csv('./data_2009/2009_p_breviews')
n_reviews_2009 = pd.read_csv('./data_2009/2009_n_reviews')
p_reviews_2009 = pd.read_csv('./data_2009/2009_p_reviews')
audience_cnt_2009 = pd.read_csv('./data_2009/2009_audience_cnt')
b_reviews_2009 = pd.merge(p_breviews_2009,n_breviews_2009)
reviews_2009 = pd.merge(p_reviews_2009,n_reviews_2009)
b_reviews_2009['breviews'] = b_reviews_2009['p_breviews'] + b_reviews_2009['n_breviews']
reviews_2009['reviews'] = reviews_2009['p_reviews'] + reviews_2009['n_reviews']
df1 = pd.merge(b_reviews_2009,reviews_2009)
online_data_2009 = pd.merge(grade_2009,df1)
online_data_2009 = pd.merge(online_data_2009,audience_cnt_2009)
online_data_2009 = online_data_2009.set_index('Unnamed: 0')
online_data_2009.index.name = 'movies'
online_data_2009
```

```python
online_data_2009.to_csv('./data_2009/online_data_2009')
```
