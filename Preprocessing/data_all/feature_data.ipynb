```python
import pandas as pd
```

```python
df1 = pd.read_excel('./data_all/dataset_feat.xlsx')
```

```python
df1.head()
```

```python
del df1['openDt']
del df1['salesAmt']
del df1['Nation']
```

```python
df1.rename(columns = {'movieNm' : 'movies','Nation_attribute' : 'natinality','openDt_attribute':'season','scrnCnt':'screen'},inplace=True)
```

```python
df1 = df1.drop_duplicates(['movies'],keep='last')
```

```python
df1
```

```python
rating = pd.read_csv('./data_all/rating_all')
```

```python
del rating['Unnamed: 0']
```

```python
df2 = pd.merge(rating,df1, on='movies')
```

```python
pactor = pd.read_csv('./data_all/pactor_all')
```

```python
del pactor['Unnamed: 0']
```

```python
df3 = pd.merge(df2,pactor, on='movies')
```

```python
pdistributor = pd.read_csv('./data_all/pdistributor_all')
```

```python
del pdistributor ['Unnamed: 0']
```

```python
df4 = pd.merge(df3,pdistributor, on='movies')
```

```python
df4.drop_duplicates(['movies'],keep='last')
```

```python
df = df4
```

```python
df.to_csv('./data_all/feature_data')
```
