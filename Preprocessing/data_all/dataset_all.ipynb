```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('./data_all/dataset_completed')
df
```

```python
del df['Unnamed: 0']
```

```python
df.set_index('movies',inplace=True)
df
```

```python
X1 = df[ df.columns[0:7] ]
X1
```

```python
X2 = df[df.columns[8:14]]
X2
```

```python
y = df['audience_cnt']
```

```python
def func(x) :
    if 3000000 > int(x) :
        return 1
   
    if 3000000 <= int(x) < 5000000 :
        return 2
     
    if 5000000 <= int(x):
        return 3
```

```python
y = y.apply(func)
y
```

```python
data1 = pd.concat([X1,y], axis=1)
```

```python
data2 = pd.concat([X2,y], axis=1)
```

```python
sns.pairplot(data1, hue='audience_cnt')
```

```python
sns.pairplot(data2, hue='audience_cnt')
```
