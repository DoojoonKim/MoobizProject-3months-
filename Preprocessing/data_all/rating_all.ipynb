```python
import pandas as pd
```

```python
df1 = pd.read_csv('./data_2017/2017_rating')
```

```python
df2 = pd.read_csv('./data_2016/2016_rating')
```

```python
df3 = pd.read_csv('./data_2015/2015_rating')
```

```python
df4 = pd.read_csv('./data_2014/2014_rating')
```

```python
df5 = pd.read_csv('./data_2013/2013_rating')
```

```python
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
df.set_index('movies',inplace=True)
```
