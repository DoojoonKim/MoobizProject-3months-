```python
import json
import math
import requests
import pandas
from pandas import datetime, Series
from urllib.request import urlopen
from tkinter import *
from urllib.parse import urlencode
import os
import pandas as pd
```

```python
df_DailySet = pd.read_excel('./data/DailyData_set(2008-2017)good.xlsx')
df_NationSet = pd.read_excel('./data/Nation_att.xlsx')
df_openDtSet = pd.read_excel('./data/openDt_att.xlsx')
```

```python
df_DailySet = df_DailySet.drop_duplicates(subset = 'movieNm', keep='first')
df_DailySet
```

```python
df_DailySet.index = pd.RangeIndex(len(df_DailySet))
df_DailySet
```

```python
df_dataset = pd.merge(df_DailySet, df_NationSet, on='movieNm')
```

```python
df_dataset = pd.merge(df_dataset, df_openDtSet, on='movieNm')
df_dataset
```
