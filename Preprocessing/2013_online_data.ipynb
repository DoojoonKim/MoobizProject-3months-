```python
import pandas as pd
grade_2013 = pd.read_csv('./data_2013/2013_grade')
n_breviews_2013 = pd.read_csv('./data_2013/2013_n_breviews')
p_breviews_2013 = pd.read_csv('./data_2013/2013_p_breviews')
n_reviews_2013 = pd.read_csv('./data_2013/2013_n_reviews')
p_reviews_2013 = pd.read_csv('./data_2013/2013_p_reviews')
audience_cnt_2013 = pd.read_csv('./data_2013/2013_audience_cnt')
b_reviews_2013 = pd.merge(p_breviews_2013,n_breviews_2013)
reviews_2013 = pd.merge(p_reviews_2013,n_reviews_2013)
b_reviews_2013['breviews'] = b_reviews_2013['p_breviews'] + b_reviews_2013['n_breviews']
reviews_2013['reviews'] = reviews_2013['p_reviews'] + reviews_2013['n_reviews']
df1 = pd.merge(b_reviews_2013,reviews_2013)
online_data_2013 = pd.merge(grade_2013,df1)
online_data_2013 = pd.merge(online_data_2013,audience_cnt_2013)
online_data_2013 = online_data_2013.set_index('Unnamed: 0')
online_data_2013.index.name = 'movies'
online_data_2013
```
