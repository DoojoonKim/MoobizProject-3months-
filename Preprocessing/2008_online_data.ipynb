```python
import pandas as pd
grade_2008 = pd.read_csv('./data_2008/2008_grade')
n_breviews_2008 = pd.read_csv('./data_2008/2008_n_breviews')
p_breviews_2008 = pd.read_csv('./data_2008/2008_p_breviews')
n_reviews_2008 = pd.read_csv('./data_2008/2008_n_reviews')
p_reviews_2008 = pd.read_csv('./data_2008/2008_p_reviews')
audience_cnt_2008 = pd.read_csv('./data_2008/2008_audience_cnt')
b_reviews_2008 = pd.merge(p_breviews_2008,n_breviews_2008)
reviews_2008 = pd.merge(p_reviews_2008,n_reviews_2008)
b_reviews_2008['breviews'] = b_reviews_2008['p_breviews'] + b_reviews_2008['n_breviews']
reviews_2008['reviews'] = reviews_2008['p_reviews'] + reviews_2008['n_reviews']
df1 = pd.merge(b_reviews_2008,reviews_2008)
online_data_2008 = pd.merge(grade_2008,df1)
online_data_2008 = pd.merge(online_data_2008,audience_cnt_2008)
online_data_2008 = online_data_2008.set_index('Unnamed: 0')
online_data_2008.index.name = 'movies'
online_data_2008
```

```python
online_data_2008.to_csv('./data_2008/online_data_2008')
```
