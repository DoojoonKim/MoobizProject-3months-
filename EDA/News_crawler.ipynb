```python
movie_news = {}
```

```python
driver = wd.Chrome('./data/chromedriver.exe')
driver.get( 'https://www.kinds.or.kr/' )
for keyword, date in movieList.items() :
    driver.find_element_by_css_selector('#searchTxt').click()
    time.sleep(2)
    driver.find_element_by_css_selector('#detailSearch_li > a > div.autoBtnImg').click()
    time.sleep(2)
    driver.find_element_by_css_selector('#li_period_all').click()
    driver.find_element_by_css_selector('#searchDetailTxt1').send_keys('%s' % keyword )
    driver.find_element_by_css_selector('#needDetailTxt1').send_keys('영화')
    driver.find_element_by_css_selector('#todate').send_keys('%s' % date[0])
    driver.find_element_by_css_selector('#fromdate').send_keys('%s' % date[1])
    driver.find_element_by_css_selector('#searchSubmit').click()
    news_cnt = driver.find_element_by_css_selector('#resultExpression > b').text
    movie_news[ keyword ] = news_cnt
    driver.get( 'https://www.kinds.or.kr/' )
```
