```python
# import urllib3
import urllib.request
import json
import pandas as pd
```

```python
req = urllib.request
```

```python
url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=ca3f4432da66e0556de8179d18762a19&openDt=2013&itemPerPage=100'
```

```python
moviePage = req.urlopen(url)
```

```python
movieData = json.loads(moviePage.read())
```

```python
movieData
```

```python
print(movieData['movieListResult']['movieList'][4]['directors'][0]['peopleNm'])
```

```python
movieDDF = pd.DataFrame()
movieDDF = movieDDF.append(
    {'영화 코드':'', '영화 제목':'', '영화제목_영어':'', '개봉일':'', '감독':''}, ignore_index = True)
```

```python
num = len(movieData['movieListResult']['movieList'])
```

```python
for i in range(0, num):
    movieDDF.loc[i,"영화 코드"] = movieData['movieListResult']['movieList'][i]['movieCd']
    movieDDF.loc[i,'영화 제목'] = movieData['movieListResult']['movieList'][i]['movieNm']
    movieDDF.loc[i,'영화 제목_영어'] = movieData['movieListResult']['movieList'][i]['movieNmEn']
    movieDDF.loc[i,'개봉일'] = movieData['movieListResult']['movieList'][i]['openDt']
    if movieData['movieListResult']['movieList'][i]['directors']!=[]:
        movieDDF.loc[i,'감독'] = movieData['movieListResult']['movieList'][i]['directors'][0]['peopleNm']
    else:
        movieDDF.loc[i,'감독'] = ''
    
movieDDF.head(10)
```

```python
movieDDF['년도'] = movieDDF['개봉일'].apply(lambda x : x[0:4])
movieDDF['월'] = movieDDF['개봉일'].apply(lambda x : x[4:6])
movieDDF['일'] = movieDDF['개봉일'].apply(lambda x : x[6:8])
```

```python
movieDDF.head(10)
```

```python
movieDDF.index = movieDDF['영화 코드']
```

```python
movieDDF
```

```python
movieDDf.to_csv('movielist.csv', index=False)
```
