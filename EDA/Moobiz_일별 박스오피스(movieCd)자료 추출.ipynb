```python
import os
import pandas as pd
```

# 일별 박스오피스 데이터셋에서 movieCd만 뽑아내기

```python
dir = os.listdir('./data')
dir.sort()
xlsx_list = [ file for file in dir if file.startswith('DailyBoxoffice_')]
```

```python
xlsx_list
```

```python
for i, data in enumerate(xlsx_list):
    if i == 0:
        df = pd.read_excel('./data/'+data)
        
    else:
        small_df = pd.read_excel('./data/'+data)
        df = pd.concat([df, small_df]).reset_index(drop = True)
        
```

```python
df.drop
```

```python
df = df.drop_duplicates(subset='movieNm', keep='first')
```

```python
df.index = pd.RangeIndex(len(df))
df
```

```python
movieCd_list = [file for file in df['movieCd']]
movieCd_list
```

```python
df_movieCd = pd.DataFrame(movieCd_list)
```

```python
df_movieCd.columns = ['movieCd']
df_movieCd
```

```python
df_movieCd.to_excel('movieCd(2013_2017).xlsx', encoding='utf-8')
```
