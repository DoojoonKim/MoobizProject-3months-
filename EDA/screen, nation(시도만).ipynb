```python
for idx, a in enumerate(df['개봉일']):
    if type(a) is not datetime.datetime:
        continue
    if a.month >= 3 and a.month <= 5: 
        df.iloc[idx]['개봉일 특성'] = 1
    elif a.month >= 6 and a.month <= 8: 
        df.iloc[idx]['개봉일 특성'] = 2
    elif a.month >= 9 and a.month <= 11: 
        df.iloc[idx]['개봉일 특성'] = 3
    else: 
        df.iloc[idx]['개봉일 특성'] = 4
df
```
