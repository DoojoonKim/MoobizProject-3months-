```python
import pandas as pd
import os
import sys
```

```python
dir = os.listdir('./data')
xlsx_list = [file for file in dir if file.startswith('BoxOffice')]
xlsx_list.sort()
```

```python
for i, data in enumerate(xlsx_list):
    if i == 0:
        df = pd.read_excel('./data/'+data)
    else:
        small_df = pd.read_excel('./data/'+data)
        df = pd.concat([df, small_df])
```

```python
df.rename(columns={'기간별 박스오피스 검색 리스트':'rank', 'Unnamed: 1':'movieNm', 'Unnamed: 2': 'openDt',
                   'Unnamed: 3':'sales', 'Unnamed: 4':'salesShare', 'Unnamed: 5':'salesAmt',
                   'Unnamed: 6':'audiCnt', 'Unnamed: 7':'audiCntAmt', 'Unnamed: 8':'scrnCnt',
                   'Unnamed: 9':'showCnt', 'Unnamed: 10':'Nation', 'Unnamed: 12':'company',
                   'Unnamed: 13':'distributor','Unnamed: 14':'rating', 'Unnamed: 15':'genre',
                   'Unnamed: 16':'director', 'Unnamed: 17':'actor'}, inplace=True)
df = df.drop('Unnamed: 11', 1)
df = df.drop(0,0)
df = df.drop(1,0)
df = df.drop(2,0)
df.index = pd.RangeIndex(len(df))
```
