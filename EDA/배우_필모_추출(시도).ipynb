```python
def actorCode(peopleNm):
#     peopleNm = '송강호'
    req = urllib.request
    url = parse.urlparse('http://www.kobis.or.kr/kobisopenapi/webservice/rest/people/searchPeopleList.json?key=%s&peopleNm=%s'%(key, peopleNm))
    query = parse.parse_qs(url.query)
    option = parse.urlencode(query, encoding='UTF-8', doseq=True)
    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/people/searchPeopleList.json?%s'%option
    peoplePage = req.urlopen(url)
    peopleData = json.loads(peoplePage.read())
    peopleSearch = peopleData['peopleListResult']['peopleList']
    for i in range(0,len(peopleSearch)):
        if peopleSearch[i]['repRoleNm'] == '배우':
            peopleCd = peopleSearch[i]['peopleCd']
            break
        elif peopleSearch[i]['repRoleNm'] == '감독':
            peopleCd = peopleSearch[i]['peopleCd']
    return peopleCd

# 배우 송강호 = 10037018
```

```python
# 배우의 필모 추출
def actFilmos(peopleCd):
    req = urllib.request
    url = parse.urlparse('http://www.kobis.or.kr/kobisopenapi/webservice/rest/people/searchPeopleInfo.json?key=%s&peopleCd=%s'%(key1, peopleCd))
    query = parse.parse_qs(url.query)
    option = parse.urlencode(query, encoding='UTF-8', doseq=True)
    url = 'http://www.kobis.or.kr/kobisopenapi/webservice/rest/people/searchPeopleInfo.json?%s'%option
    peoplePage = req.urlopen(url)
    peopleData = json.loads(peoplePage.read())
    # print(peopleData)
    actorFilmos = peopleData['peopleInfoResult']['peopleInfo']['filmos']
    actorFilmoMovie = {}
    for i in range(0,len(actorFilmos)):
        actOrNot = actorFilmos[i]['moviePartNm']
        if actOrNot == '배우':
            actorFilmoMovie[actorFilmos[i]['movieCd']] =  actorFilmos[i]['movieNm']
    return actorFilmoMovie
```
